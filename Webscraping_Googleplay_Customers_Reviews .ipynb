Webscraping Reviews from Googleplay store
-----------

Below is the code chunk used in the extraction of `Urgent10k` and `Irorun` customers' reviews from the google playstore.
Data source:
- `Irorun` - [Google Play Store](https://play.google.com/store/apps/details?id=com.irorun)
- `Urgent10k` - [Google Play Store](https://play.google.com/store/apps/details?id=ng.urgent10k)

- google app id = ng.urgent10k
- google app id = com.irorun

Web scraped data on: 2023/03/20

```python
# If you haven't intsalled google_play-scraper in your environment, remove '#' before pip and run
# pip install google-play-scraper
```

```python
# import all necessary libraries
import numpy as np
import pandas as pd
from google_play_scraper import Sort, reviews_all, reviews, app
```

```python
# scrape all reviews --'change the app id from google within the quote'---
result = reviews_all(
    'ng.urgent10k', # the app id
    sleep_milliseconds=0, 
    lang='en',            # language'en' 
    country='us',         # USA English
    sort=Sort.NEWEST,   
)
```

```python
# convert scraped result into a dataframe 
scrapeddata = pd.DataFrame(np.array(result),columns=['review'])
scrapeddata = scrapeddata.join(pd.DataFrame(scrapeddata.pop('review').tolist()))
scrapeddata.head()
```

```python
# count the data we got
len(scrapeddata.index)
```

```python
# Filter the scrapped data
scrapeddata[['content','score','at']].head()
```

```python
# code to sort the date(if you want to)
scrappeddata1 = scrapeddata[['content','score','at']]
sorteddata = scrappeddata1.sort_values(by='at', ascending=False) # sort in descending order
sorteddata.head()
```

```python
# export the scrapped data
scrapeddata.to_csv("Urgent10k_reviews.csv", index = False) 
```
